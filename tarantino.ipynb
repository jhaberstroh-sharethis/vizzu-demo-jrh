```python
# Import components

import pandas as pd
from ipyvizzu import Chart, Data, Config, Style
from ipyvizzustory import Story, Slide, Step
```

```python
# Tarantino Profane Words and Deaths by 538 Data Set 

data_frame3 = pd.read_csv('https://raw.githubusercontent.com/vizzuhq/vizzu-workshops/main/2023-04-25-PyData-Seattle/data/tarantino.csv',)
display(data_frame3)
```

```python
# Add one of the data frames to the ipyvizzu-story Data object

df_to_use = data_frame3.copy()
df_to_use['count'] = 1
df_to_use['minutes_in_round'] = ((df_to_use['minutes_in']/10).round()*10).astype(int).astype(str)
keep_words = df_to_use.groupby('word').agg(qty=pd.NamedAgg('word','count')).reset_index().query('qty > 40')['word'].values
df_to_use = df_to_use[df_to_use.word.isin(keep_words)]

data = Data()
data.add_data_frame(df_to_use) #data_frame2

## movie	type	word	minutes_in
## 0	Reservoir Dogs	word	dick	0.40
## 1	Reservoir Dogs	word	dicks	0.43
## 2	Reservoir Dogs	word	fucked	0.55
## 3	Reservoir Dogs	word	fucking	0.61
## 4	Reservoir Dogs	word	bullshit	0.61
```

```python
print(df_to_use)
```

```python
# Set the style of the charts in the story

style = Style({ 
            'plot': { 
                'marker': { 'label' :{ 
                    'numberFormat' : 'prefixed',
                    'maxFractionDigits' : '1',
                    'numberScale':'shortScaleSymbolUS'
                }},
                     'xAxis': { 'label': { 'angle': '-45deg',  'numberScale':'shortScaleSymbolUS'}},
                     'yAxis': { 'label': { 'numberScale':'shortScaleSymbolUS'}} 
                },
                'legend' : {'width' : '15%'},
        })
```

```python
# Create Story object, set size

story = Story(data=data,style=style)
story.set_feature('tooltip', True)
story.set_size('100%', '400px')
```

```python
slide1 = Slide(
    Step(
        Config.groupedBar({ 
                'y': 'word',
                'x': 'count',
                'title': 'curses',
                'groupedBy': 'word',
            })
))

story.add_slide(slide1)

slide2 = Slide(
    Step(
        Config.stackedColumn({ #stackedColumn (stackedBy), groupedColumn (groupedBy), percentageArea (stackedBy), stream (stackedBy), violin (splittedBy)
                'x': 'minutes_in_round',
                'y': 'count',
                'stackedBy': 'word',
                'title': '',
        }),
))

story.add_slide(slide2)

story.play()
```
